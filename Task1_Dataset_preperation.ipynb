```python
import pandas as pd
import re
import string

# Load the data
file_path = "/kaggle/input/d/tanmaypawar/daily-dialouge/data.csv"
test_fpath = "/kaggle/input/validation-data/test_data.csv"
data = pd.read_csv(test_fpath)

def clean_text(text):
    # Remove leading and trailing whitespaces
    text = text.strip()
    # Remove extra spaces between words
    text = re.sub(r'\s+', ' ', text)
    # Fix spacing around punctuation
    text = re.sub(r'\s+([?.!,])', r'\1', text)
    # Convert to lowercase
    text = text.lower()
    # Remove non-ASCII characters
    text = text.encode("ascii", "ignore").decode()
    return text

# Apply cleaning function to each conversation
data["Cleaned_Conversation"] = data["Conversation"].apply(clean_text)

# Drop duplicates and NaN values
data.drop_duplicates(subset="Cleaned_Conversation", inplace=True)
data.dropna(subset=["Cleaned_Conversation"], inplace=True)

# Save cleaned data to a new CSV file
output_path = "cleaned_data.csv"
data.to_csv(output_path, index=False)

print(f"Data cleaning completed. Cleaned data saved to {output_path}.")
```

```python
import subprocess
import threading
import time

# Function to start Ollama server in the background
def run_kar_de():
  def start_ollama():
    subprocess.run(["ollama", "serve"])

# Run Ollama in a daemon thread
  ollama_thread = threading.Thread(target=start_ollama, daemon=True)
  ollama_thread.start()

  # Wait a few seconds for server to start
  time.sleep(5)

  print("✅ Ollama server is running in the background!")
```

```python
run_kar_de()

import os
import subprocess

# Create the output directory if it doesn't exist
output_dir = "outputs"
os.makedirs(output_dir, exist_ok=True)
# Extended few-shot examples for better prompt engineering
examples = [
    ("Dialogue: 'Can we reschedule the meeting to next Monday? I have another appointment.' "
     "Summary: 'Request to reschedule meeting due to a prior commitment.'"),
    ("Dialogue: 'I can't believe I forgot to bring my keys again. This is so frustrating.' "
     "Summary: 'Person A is frustrated about forgetting keys repeatedly.'"),
    ("Dialogue: 'The movie was great! I really loved the plot twist at the end.' "
     "Summary: 'Person A enjoyed the movie, especially the plot twist.'"),
    ("Dialogue: 'I’m thinking of trying a new recipe tonight. Maybe something Italian.' "
     "Summary: 'Person A plans to cook a new Italian dish.'"),
    ("Dialogue: 'I’ve been so stressed lately with work and deadlines piling up.' "
     "Summary: 'Person A feels overwhelmed by work and deadlines.'"),
    ("Dialogue: 'Can you help me with my math homework? I’m stuck on these equations.' "
     "Summary: 'Person A seeks assistance with math homework.'"),
    ("Dialogue: 'I just got promoted! I’m so excited and a bit nervous about the new responsibilities.' "
     "Summary: 'Person A is happy about a promotion but nervous about new duties.'"),
    ("Dialogue: 'The weather has been awful lately. It’s been raining nonstop.' "
     "Summary: 'Person A complains about continuous rainy weather.'"),
    ("Dialogue: 'I really appreciate your help with the project. I couldn’t have done it without you.' "
     "Summary: 'Person A expresses gratitude for help with a project.'")
]

# Combine examples into a single prompt
example_prompt = "\n".join([f"Example {i+1}: {ex}" for i, ex in enumerate(examples)])

# Array of conversation snippets
prompts = data

# Loop through prompts
for idx, conversation in enumerate(prompts):
    # Construct the enhanced prompt
    full_prompt = (
        f"{example_prompt}\n\n"
        f"Summarize the following conversation concisely and accurately. "
        f"Keep the summary length within 70% of the dialogue length(But don't write how much reduction is added in the output)."
        f"The summary format must be like Here is the  summary: The summary generated."
        f"Avoid unnecessary details and focus on the main points:\n{conversation}"
    )

    # Define the output file path
    output_file = os.path.join(output_dir, f"{idx}.txt")

    # Run Ollama LLaMA 3 model with hyperparameter tuning
    result = subprocess.run(
        ["ollama", "run", "llama3", full_prompt],
        capture_output=True, text=True
    )

    # Write the output to a file
    with open(output_file, "w", encoding="utf-8") as f:
        f.write(result.stdout)

    print(f"✅ Summary saved: {output_file}")
```

```python
import os
import pandas as pd

# Define the directory where output files are stored
output_dir = "outputs"

# Get all .txt files sorted by index (filename)
txt_files = sorted(
    [f for f in os.listdir(output_dir) if f.endswith(".txt")],
    key=lambda x: int(os.path.splitext(x)[0])  # Sort numerically by filename
)

# Read each file's content and store in a list
summaries = []
for file in txt_files:
    file_path = os.path.join(output_dir, file)
    with open(file_path, "r", encoding="utf-8") as f:
        summaries.append(f.read().strip())  # Read and remove extra spaces/newlines

# Create a DataFrame with the new column
df_summaries = pd.DataFrame({"Summaries": summaries})

# Assume `df` is the original DataFrame (Ensure it has the same number of rows)
df2["Summaries"] = df_summaries["Summaries"]

# Save the merged DataFrame as a CSV file
df2.to_csv("final_output.csv", index=False)

print("✅ Merged output saved as 'final_output.csv'")
```
